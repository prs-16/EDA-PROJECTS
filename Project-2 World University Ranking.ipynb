# Exploratory Analysis and  Vizualization of Data

```python
import seaborn as sns
import matplotlib
import matplotlib.pyplot as plt
%matplotlib inline

sns.set_style('darkgrid')
matplotlib.rcParams['font.size'] = 14
matplotlib.rcParams['figure.figsize'] = (9, 5)
matplotlib.rcParams['figure.facecolor'] = '#00000000'
```

```python
top_countries = university_df.country.value_counts().head(15)
top_countries
```

```python
plt.figure(figsize=(12,6))
plt.xticks(rotation=-90)
sns.barplot(x = top_countries.index,y = top_countries)
```

```python
publ = university_df.publications.head(15)
publ
```

```python
World_Rank = university_df.world_rank.head(15)
World_Rank
```

```python
quality = university_df.quality_of_education.head(15)
quality
```

```python
plt.figure(figsize=(12,6))
plt.xticks(rotation=90)
sns.barplot(x = World_Rank,y = publ)
```

```python
plt.figure(figsize=(12,6))
plt.xticks(rotation=-90)
sns.barplot(x = World_Rank,y = quality)
```

## Comparing the universities according to their score with respective year

```python
plt.figure(figsize=(12,6))
plt.xticks(rotation=90)
sns.barplot(x = university_df['year'],y = university_df['score'])
```

# Comparing the universities according to their score with respective year using line plot

```python
university_df_2012 = university_df[university_df.year==2012]
university_df_2012.head()
```

```python
university_2012 = university_df_2012.head(20)
plt.plot(university_2012.country,university_2012.institution,)
plt.scatter(university_2012.country,university_2012.institution,)
plt.xlabel('countries')
plt.ylabel('universities')
plt.title('Rank in 2012')
plt.show()
```

```python
university_df_2013 = university_df[university_df.year==2013]
university_df_2013.head()
university_2013 = university_df_2013.head(20)
plt.plot(university_2013.country,university_2013.institution,)
plt.scatter(university_2013.country,university_2013.institution,)
plt.xlabel('countries')
plt.ylabel('universities')
plt.title('Rank in 2013')
plt.show()
```

```python
university_df_2014 = university_df[university_df.year==2014]
university_df_2014.head()
university_2014 = university_df_2014.head(20)
plt.plot(university_2014.country,university_2014.institution,)
plt.scatter(university_2014.country,university_2014.institution,)
plt.xlabel('countries')
plt.ylabel('universities')
plt.title('Rank in 2014')
plt.show()
```

```python
university_df_2015 = university_df[university_df.year==2015]
university_df_2015.head()
university_2015 = university_df_2015.head(20)
plt.plot(university_2015.country,university_2015.institution,)
plt.scatter(university_2015.country,university_2015.institution,)
plt.xlabel('countries')
plt.ylabel('universities')
plt.title('Rank in 2015')
plt.show()
```

```python
plt.figure(figsize=(8,8))
sns.set_style("darkgrid")

plt.plot(university_2012.country,university_2012.institution,'s-')
plt.plot(university_2013.country,university_2013.institution,'o--')
plt.plot(university_2014.country,university_2014.institution,'*--')
plt.plot(university_2015.country,university_2015.institution,'--')
plt.xlabel('Countries')
plt.ylabel('Universities')
plt.legend(['Un_2012', 'Un_2013','Un_2014','Un_2015']);
```
